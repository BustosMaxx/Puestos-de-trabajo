```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, time, datetime

puesto_trabajo = pd.read_csv("Puestos-de-trabajo/puestos_depto_total_por_clae2.csv", na_values=[-99])
dic_depto = pd.read_csv("Puestos-de-trabajo/diccionario_cod_depto.csv")
dic_clae2 = pd.read_csv("Puestos-de-trabajo/diccionario_clae2.csv")
trabajo_publico = pd.read_csv("Puestos-de-trabajo/puestos_depto_pub.csv", sep=";", encoding="ANSI", na_values=[-99])
trabajo_privado = pd.read_csv("Puestos-de-trabajo/puestos_depto_priv.csv", sep=";", na_values=[-99])

puesto_trabajo.dropna(inplace=True, subset=["id_provincia_indec"])
```

```python
''' Se agrega columna de fecha con el año y el mes. ''' 

times = [datetime.strptime(i, '%Y-%m-%d') for i in puesto_trabajo.fecha]
puesto_trabajo["año"] = [times[i].year for i in range(0,len(puesto_trabajo))]
puesto_trabajo["mes"] = [times[i].month for i in range(0,len(puesto_trabajo))]
puesto_trabajo = puesto_trabajo[puesto_trabajo.mes.isin([12])]
puesto_trabajo.head()
```

```python
times = [datetime.strptime(str(trabajo_publico.iloc[i,0]), '%d/%m/%Y') for i in np.arange(len(trabajo_publico))]

trabajo_publico["año"] = [times[i].year for i in range(0,len(trabajo_publico))]
trabajo_publico["mes"] = [times[i].month for i in range(0,len(trabajo_publico))]

trabajo_publico = trabajo_publico[trabajo_publico.mes.isin([12])]
trabajo_publico.head()
```

```python
times = [datetime.strptime(str(trabajo_privado.iloc[i,0]), '%d/%m/%Y') for i in np.arange(len(trabajo_privado))]

trabajo_privado["año"] = [times[i].year for i in range(0,len(trabajo_privado))]
trabajo_privado["mes"] = [times[i].month for i in range(0,len(trabajo_privado))]

trabajo_privado = trabajo_privado[trabajo_privado.mes.isin([12])]
```

```python
trabajo_pub_priv = pd.concat([trabajo_publico,trabajo_privado])
trabajo_pub_priv = trabajo_pub_priv.groupby(by=["año","sector"], as_index=False)["puestos"].sum()
```

```python
trabajo_pub_priv.groupby("año")["puestos"].sum()
```

```python
sns.set_theme(style="whitegrid")

# Draw a nested barplot 
g = sns.catplot(
    data=trabajo_pub_priv, kind="bar",
    x="año", y="puestos", hue="sector",
    #errorbar="sd",
    #palette="dark", 
    #alpha=.6, 
    #height=6
)
g.despine(left=True)
g.set_axis_labels("", "# de puestos")
g.legend.set_title("")
```

## Provincia

```python
rubros_cantidad_prov = rubros.groupby(["año","id_provincia_indec"], as_index=False)["puestos"].sum()
rubros_cantidad_prov = rubros_cantidad_prov.sort_values("puestos", ascending = False)
prov_top_ten =pd.merge(rubros_cantidad_prov, dic_prov, how= "inner", on="id_provincia_indec")

prov2021 = prov_top_ten[prov_top_ten["año"] == 2021]
```

```python
data = prov2021.copy()
data = data.head(10)

labels = list(data["nombre_provincia_indec"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid",
             font_scale=1.2)

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(8, 8))

# Load the example car crash dataset
#data = sns.load_dataset("prov2021").sort_values("puestos", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="puestos", y="nombre_provincia_indec", data=data,
            label="nombre_provincia_indec", 
            color="b"
           )

# Add a legend and informative axis label
#ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="N° de empleados por provincia - 2021",
       #xlim=(0, 24),
      )
sns.despine(left=True, bottom=True)
```

## Histograma por año

```python
rubros_cantidad_año = rubros.groupby(["año"], as_index=False)["puestos"].sum()

sns.set_theme(style="whitegrid")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=rubros_cantidad_año, kind="bar",
    x="año", y="puestos",
    #hue="sex", 
    #errorbar="sd", 
    palette="dark", 
    alpha=.6, 
    height=6
)
g.despine(left=True)
g.set_axis_labels("año", "Cantidad de puestos")
#g.legend.set_title("")
```

# Headmap

## Provincia vs año

```python
prov_top_ten["puestos"] = prov_top_ten["puestos"].astype("int64")
data = prov_top_ten.pivot("nombre_provincia_indec","año","puestos")


''' Mapa de calor'''
f, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(data=data, annot=True, fmt="d", linewidths=0.5, ax=ax)
ax.set(ylabel="",
       xlabel="año",
       #xlim=(0, 24),
      )
```

## Provincia vs año relativos a su población.

```python
poblacion = pd.read_excel("Puestos-de-trabajo/Población2010.xls")
poblacion = poblacion.rename(columns={"Provincia":"nombre_provincia_indec"})
poblacion_provincia = pd.merge(poblacion, dic_prov, how= "inner", on="nombre_provincia_indec" )

gb_df = prov_top_ten.copy()
gb_df["%"]= np.zeros(len(gb_df))

test = pd.merge(gb_df,poblacion, how="inner", on="nombre_provincia_indec")
test["%"] = np.round((test ["puestos"]/test["Población"])*100,2).astype("int64")

data_relativa = test.pivot("nombre_provincia_indec","año","%")
```

### Mapa de calor

```python
f, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(data=data_relativa, annot=True, fmt="d", linewidths=0.5, ax=ax)
ax.set(ylabel="",
       xlabel="año",
       #xlim=(0, 24),
      )
```
